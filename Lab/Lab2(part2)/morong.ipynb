# 1. Làm thế nào để đánh giá chất lượng của các cạnh được phát hiện bởi Canny?
- chúng ta có thể so sánh nó với ảnh gốc để đánh giá :
    + biên có xuất hiện đúng vị trí ứng với ảnh gốc hay không.
    + biên có bị nhiễu ở vùng phẳng không.
    + biên có liền mạch hay không.
    + biên mảnh hay dày.
- hoặc chúng ta có thể so sánh nó với các thuật toán tìm biên như Sobel, Laplacian, .... nếu Canny ít nhiễu hơn, biên mảnh hơn, Biên rõ và liên tục hơn. thì canny cho kết quả Chất lượng cạnh tốt hơn.
- => để đánh giá chất lượng của các cạnh được phát hiện bởi Canny thì hãy đánh giá độ chính xác vị trí, mức nhiễu, độ mảnh và tính liên tục của các cạnh 


# 2. Có những phương pháp nào khác để cải thiện hiệu suất của Canny?
- Cải thiện bước tiền xử lý : đầu vào tốt thì kết quả cũng tốt, Việc giảm nhiễu ảnh trước khi áp dụng Canny có ảnh hưởng trực tiếp đến chất lượng các cạnh được phát hiện. Có thể sử dụng các bộ lọc như Gaussian, Median hoặc Bilateral để giảm nhiễu nhưng vẫn giữ được biên.
-  Điều chỉnh tham số của thuật toán: 
    + Tham số σ (sigma) trong bộ lọc Gaussian ảnh hưởng đến mức độ làm mờ ảnh. Sigma lớn giúp giảm nhiễu tốt hơn nhưng có thể làm mất chi tiết biên; sigma nhỏ giữ chi tiết tốt hơn nhưng dễ bị nhiễu.
    + Ngưỡng thấp và ngưỡng cao ảnh hưởng đến việc giữ hay loại bỏ các cạnh yếu. Việc chọn ngưỡng phù hợp giúp cân bằng giữa độ đầy đủ và độ chính xác của cạnh.
- chọn ngưỡng dựa vào histogram ảnh. Phân tích phân bố mức xám hoặc gradient trong ảnh để xác định các ngưỡng phù hợp thay vì chọn thủ công.
- sử dụng các phương pháp chọn ngưỡng tự động như Median-based threshold 


# 3. Canny có thể được sử dụng để phát hiện các cạnh trong ảnh màu không? Nếu có, thì như thế nào?
- Canny có thể sử dụng để phát các cạnh trong ảnh màu nhưng cânf phải xử lý ảnh trước khi áp dụng canny.

- có thể chuyển ảnh màu qua ảnh xám cách này đơn giản, nhanh và rất phổ biến. nhưng có thể mất biên do khác biệt màu nhưng độ sáng giống nhau
- có thể Áp dụng Canny trên từng kênh màu (R, G, B) ,Sau đó kết hợp các biên lại .Ưu điểm là phát hiện được biên do thay đổi màu sắc, nhược điểm nhiễu nhiều hơn ,Tốn tính toán hơn.
- Chuyển sang không gian màu khác (HSV, Lab). cách này cần tách độ sáng (L hoặc V) khỏi thông tin màu sau đó áp dụng Canny lên kênh độ sáng. cách này tốt nhất nó biên tốt, ổn định với thay đổi độ sáng nhưng mà phức tạp hơn so với chuyển sang ảnh xám trực tiếp.


# 4. Làm thế nào để áp dụng Canny cho các video?
- Để áp dụng Canny cho video, ta xử lý video như một chuỗi các khung hình tĩnh. Mỗi frame được chuyển sang ảnh xám, giảm nhiễu, sau đó áp dụng thuật toán Canny. Các kết quả biên của từng frame được ghép lại để tạo thành video biên.
