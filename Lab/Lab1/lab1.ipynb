```python
import cv2
from PIL import Image
from matplotlib import pyplot as plt
import numpy as np

print(cv2.__version__)
print(Image.__version__)
```

```python
# đọc ảnh bằng opencv
imgOpenCV = cv2.imread('../image/test.jpg')
cv2.imshow('OpenCV Image', imgOpenCV)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# đọc ảnh bằng pillow
imgPillow = Image.open('../image/test.jpg')
imgPillow.show()
```

```python
# lưu ảnh với định dạng khác
cv2.imwrite('../image/Opencv.png', imgOpenCV)
```

```python
imgPillow.save('../image/Pillow.png')
```

```python
# chuyển đổi sang gray
img_gray = cv2.cvtColor(imgOpenCV, cv2.COLOR_BGR2GRAY)

plt.imshow(img_gray, cmap='gray')
plt.title('Ảnh xám')
plt.axis('off')
plt.show()
```

```python
img_gray1 = imgPillow.convert('L')

plt.imshow(img_gray1, cmap='gray')
plt.title('Ảnh xám')
plt.axis('off')
plt.show()
```

```python
# chuyển đổi sang HSV, LAB
img_hsv = cv2.cvtColor(imgOpenCV, cv2.COLOR_BGR2HSV)
img_lab = cv2.cvtColor(imgOpenCV, cv2.COLOR_BGR2LAB)

fig, axes = plt.subplots(1, 2, figsize=(20, 10))
axes[0].imshow(img_hsv)
axes[0].set_title('Ảnh HSV')
axes[0].axis('off')

axes[1].imshow(img_lab)
axes[1].set_title('Ảnh LAB')
axes[1].axis('off')
plt.show()
```

```python
img_hsv1 = imgPillow.convert('HSV')
print(img_hsv1.mode)
```

```python
img_lab1 = imgPillow.convert('LAB')
print(img_lab1.mode)
```

```python
# cắt ảnh
plt.imshow(imgPillow.crop((100, 100, 200, 200)))
plt.title('Ảnh cắt')
plt.axis('off')
plt.show()
```

```python
plt.imshow(imgOpenCV[100:200, 100:200])
plt.title('Ảnh cắt')
plt.axis('off')
plt.show()
```

```python
# resize ảnh
plt.imshow(imgPillow.resize((300, 400)))
plt.title('Ảnh resize')
plt.axis('off')
plt.show()
```

```python

plt.imshow(cv2.resize(imgOpenCV,(300,300)))
plt.title('Ảnh resize')
plt.axis('off')
plt.show()
```

```python
# vẽ ảnh
imgDraw = imgOpenCV.copy()
cv2.line(imgDraw, (0, 0), (100, 100), (255, 0, 0), 5)
cv2.circle(imgDraw, (50, 50), 10, (0, 255, 0), 2)
cv2.rectangle(imgDraw, (50, 50), (200, 200), (0, 255, 0), 3)
cv2.putText(imgDraw, 'Text in openCV', (0, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

plt.imshow(imgDraw)
plt.title('Ảnh vẽ')
plt.axis('off')
plt.show()
```

```python
from PIL import ImageDraw

draw = ImageDraw.Draw(imgPillow)

# line
draw.line([(0, 0), (100, 100)], (255, 0, 0), 5)
draw.rectangle([(50, 50), (200, 200)], outline=(0, 255, 0), width=3)
draw.ellipse([(50, 50), (60, 60)], outline=(0, 255, 0), width=2)
draw.text((0, 10), "Text in Pillow", fill=(255, 0, 0), width=2)


plt.imshow(imgPillow)
plt.title('Ảnh vẽ')
plt.axis('off')
plt.show()
```
