```python
import cv2
from PIL import Image

print(cv2.__version__)
print(Image.__version__)
```

```python
# đọc ảnh bằng opencv
imgOpenCV = cv2.imread('image/sample_parrot-image.jpg')
cv2.imshow('Display with opencv', imgOpenCV)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# đọc ảnh bằng pillow
imgPillow = Image.open('image/sample_parrot-image.jpg')
imgPillow.show()
```

```python
# lưu ảnh với định dạng khác
cv2.imwrite('image/Opencv.png', imgOpenCV)
```

```python
imgPillow.save('image/Pillow.png')
```

```python
# chuyển đổi sang gray
img_gray = cv2.cvtColor(imgOpenCV, cv2.COLOR_BGR2GRAY)
cv2.imshow('Gray Image', img_gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img_gray1 = imgPillow.convert('L')
img_gray1.show()
```

```python
# chuyển đổi sang HSV, LAB
img_hsv = cv2.cvtColor(imgOpenCV, cv2.COLOR_BGR2HSV)
img_lab = cv2.cvtColor(imgOpenCV, cv2.COLOR_BGR2LAB)
cv2.imshow('HSV Image', img_hsv)
cv2.imshow('LAB Image', img_lab)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img_hsv1 = imgPillow.convert('HSV')
img_hsv1.show()
print(img_hsv1.mode)
```

```python
img_lab1 = imgPillow.convert('LAB')
img_lab1.show()
print(img_lab1.mode)
```

```python
# cắt ảnh
imgPillow.crop((100, 100, 200, 200)).show()
```

```python
imgCrop = imgOpenCV[100:200, 100:200]
cv2.imshow('Image Crop', imgCrop)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# resize ảnh
imgPillow.resize((100, 300)).show()
```

```python
imgResize = cv2.resize(imgOpenCV,(100,300))
cv2.imshow('Image resize', imgResize)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# vẽ ảnh
imgDraw = imgOpenCV.copy()
cv2.line(imgDraw, (0, 0), (100, 100), (255, 0, 0), 5)
cv2.circle(imgDraw, (50, 50), 10, (0, 255, 0), 2)
cv2.rectangle(imgDraw, (50, 50), (200, 200), (0, 255, 0), 3)
cv2.putText(imgDraw, 'Text in openCV', (0, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
cv2.imshow('Image', imgDraw)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
from PIL import ImageDraw

draw = ImageDraw.Draw(imgPillow)

# line
draw.line([(0, 0), (100, 100)], (255, 0, 0), 5)
draw.rectangle([(50, 50), (200, 200)], outline=(0, 255, 0), width=3)
draw.ellipse([(50, 50), (60, 60)], outline=(0, 255, 0), width=2)
draw.text((0, 10), "Text in Pillow", fill=(255, 0, 0))
imgPillow.show()
```

```python
imgPillow.close()
```
