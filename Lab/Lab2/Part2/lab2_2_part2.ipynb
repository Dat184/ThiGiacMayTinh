```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("../../image/noise-img.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(gray, 100, 200)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.title("nh gc")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.subplot(1,2,2)
plt.title("Canny - OpenCV")
plt.imshow(edges, cmap="gray")
plt.axis("off")

plt.show()
```

```python
from skimage import io, color, feature
import matplotlib.pyplot as plt

img = io.imread("../../image/noise-img.jpg")

gray = color.rgb2gray(img)

edges = feature.canny(gray, sigma=1.5, low_threshold=0.1, high_threshold=0.3)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.title("nh gc")
plt.imshow(img)
plt.axis("off")

plt.subplot(1,2,2)
plt.title("Canny - Scikit-image")
plt.imshow(edges, cmap="gray")
plt.axis("off")

plt.show()
```
