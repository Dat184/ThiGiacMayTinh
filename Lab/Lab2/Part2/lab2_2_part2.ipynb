```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("../../image/noise-img.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(gray, 100, 200)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.title("nh gc")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.subplot(1,2,2)
plt.title("Canny - OpenCV")
plt.imshow(edges, cmap="gray")
plt.axis("off")

plt.show()
```

```python
from skimage import io, color, feature
import matplotlib.pyplot as plt

img = io.imread("../../image/noise-img.jpg")

gray = color.rgb2gray(img)

edges = feature.canny(gray, sigma=1.5, low_threshold=0.1, high_threshold=0.3)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.title("nh gc")
plt.imshow(img)
plt.axis("off")

plt.subplot(1,2,2)
plt.title("Canny - Scikit-image")
plt.imshow(edges, cmap="gray")
plt.axis("off")

plt.show()
```

```python

edges_sigma_low = feature.canny(gray, sigma=1.0, low_threshold=0.1, high_threshold=0.3)
edges_sigma_high = feature.canny(gray, sigma=3.0, low_threshold=0.1, high_threshold=0.3)
edges_thresh_low = feature.canny(gray, sigma=2.0, low_threshold=0.05, high_threshold=0.1)
edges_thresh_high = feature.canny(gray, sigma=2.0, low_threshold=0.2, high_threshold=0.5)
# --- HIN TH KT QU ---
fig, axes = plt.subplots(2,3, figsize=(12, 10))

axes[0,0].imshow(gray, cmap='gray')
axes[0,0].set_title('nh gc')
axes[0,0].axis('off')

# Hng 1: So snh Sigma
axes[0, 1].imshow(edges_sigma_low, cmap='gray')
axes[0, 1].set_title('Sigma thp')
axes[0, 1].axis('off')

axes[0, 2].imshow(edges_sigma_high, cmap='gray')
axes[0, 2].set_title("Sigma cao")
axes[0, 2].axis('off')

# Hng 2: So snh Threshold
axes[1, 0].imshow(edges_thresh_low, cmap='gray')
axes[1, 0].set_title('Ngng Thp')
axes[1, 0].axis('off')

axes[1, 1].imshow(edges_thresh_high, cmap='gray')
axes[1, 1].set_title('Ngng Cao')
axes[1, 1].axis('off')

axes[1, 2].axis('off')

plt.tight_layout()
plt.show()
```
