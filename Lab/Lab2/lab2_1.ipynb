```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def adjust_brightness(image, brightness):
    adjusted_image = image.astype(np.int16) + brightness
    adjusted_image = np.clip(adjusted_image, 0, 255)
    adjusted_image = adjusted_image.astype(np.uint8)
    return adjusted_image
```

```python
def adjust_contrast(image, gamma):
    adjusted_image = image.astype(np.float32) / 255.0
    adjusted_image = (adjusted_image ** gamma) * 255
    adjusted_image = adjusted_image.astype(np.uint8)
    return adjusted_image
```

```python
def adjust_negative(image):
    adjusted_image = 255 - image
    return adjusted_image
```

```python
imgOpenCV = cv2.imread('../image/test.jpg')

image = cv2.cvtColor(imgOpenCV, cv2.COLOR_BGR2RGB)
image_bright = adjust_brightness(image, brightness=50)
image_contrast = adjust_contrast(image, gamma=2)
image_negative = adjust_negative(image)

fig, axes = plt.subplots(1, 4, figsize=(20, 10))

axes[0].imshow(image)
axes[0].set_title('nh gc')
axes[0].axis('off')

axes[1].imshow(image_bright)
axes[1].set_title('nh  sng')
axes[1].axis('off')

axes[2].imshow(image_contrast)
axes[2].set_title('nh tng phn')
axes[2].axis('off')

axes[3].imshow(image_negative)
axes[3].set_title('nh m bn')
axes[3].axis('off')


plt.tight_layout()
plt.show()
```
