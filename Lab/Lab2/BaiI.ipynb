```python

import cv2
import matplotlib.pyplot as plt
img = cv2.imread('../image/1234.jpg')
imgrgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(imgrgb)
plt.title("img_rgb")
plt.show()
```

```python
#BaiI
#tang giam do sang va tuong phan
plt.figure()
increase_Brightness = cv2.convertScaleAbs(imgrgb, alpha=1, beta=50)
plt.subplot(2,2,1)
plt.title("increase_Brightness")
plt.imshow(increase_Brightness)
decrease_Brightness = cv2.convertScaleAbs(imgrgb, alpha=1, beta=-50)
plt.subplot(2,2,2)
plt.title("decrease_Brightness")
plt.imshow(decrease_Brightness)
increase_contrast = cv2.convertScaleAbs(imgrgb, alpha=1.5, beta=0)
plt.subplot(2,2,3)
plt.imshow(increase_contrast)
plt.title("increase_contrast")
decrease_contrast = cv2.convertScaleAbs(imgrgb, alpha=0.5, beta=0)
plt.subplot(2,2,4)
plt.imshow(decrease_contrast)
plt.title("decrease_contrast")
plt.show()
```

```python
# bien doi am ban anh
negative_img = 255-imgrgb
plt.imshow(negative_img)
plt.title("negative_img")
plt.show()
```

```python
# Ct ngng
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
hist = cv2.calcHist([gray],[0],None,[256],[0,256])

plt.plot(hist)
plt.title("Histogram")
plt.xlabel("Pixel value")
plt.ylabel("num of pixel")
plt.show()
maxValue = 255
thresh, thresholded_img = cv2.threshold(gray, 0, maxValue, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(gray, cmap='gray')
plt.title("img")
plt.subplot(1,2,2)
plt.imshow(thresholded_img, cmap='gray')
plt.title("img_thresholded")
plt.show()
```
