```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('../image/1234.jpg')
imggray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(imggray, cmap='gray')
plt.title("img_gray")
plt.show()
```

```python
#kernel Sobel
sobelx = cv2.Sobel(imggray, cv2.CV_64F, 1, 0, ksize=3)
sobely = cv2.Sobel(imggray, cv2.CV_64F, 0, 1, ksize=3)
gradient_soble = cv2.magnitude(sobelx, sobely)
plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
plt.imshow(gradient_soble, cmap='gray')
plt.title("gradient_soble")
#kernel Prewitt
kernelx = np.array([[1,0,-1],
                    [1,0,-1],
                    [1,0,-1]])
kernely = np.array([[1,1,1],
                    [0,0,0],
                    [-1,-1,-1]])
prewittx = cv2.filter2D(imggray,cv2.CV_64F, kernelx)
prewitty = cv2.filter2D(imggray, cv2.CV_64F, kernely)
gradient_prewitt = cv2.magnitude(prewittx, prewitty)
plt.subplot(1,2,2)
plt.imshow(gradient_prewitt, cmap='gray')
plt.title("gradient_prewitt")
plt.show()
```
