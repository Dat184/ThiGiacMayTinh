```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
img = cv2.imread('../image/1234.jpg')
imgrgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(imgrgb)
plt.title("img_rgb")
plt.show()
```

```python

kernel_size = 15 #laf so ler
#Loc trung Binh
def meam_filter(img,kernel_size):
    if kernel_size % 2 ==0 :
        print("kernel_size phai la so le")
        return None
    else:
        return cv2.blur(img, (kernel_size, kernel_size))
#Loc Gaussian
def gaussian_filter(img,kernel_size):
    if kernel_size % 2 ==0 :
        print("kernel_size phai la so le")
        return None
    else:
        return cv2.GaussianBlur(img, (kernel_size, kernel_size),0)
#Lam Sac net
def sharp(img):
    laplacian = cv2.Laplacian(img, cv2.CV_64F)
    laplacian_abs = cv2.convertScaleAbs(laplacian) 
    alpha = 1.0
    sharp = cv2.addWeighted(img, 1.0, laplacian_abs, alpha, 0)
    return sharp
imggray = cv2.imread('../image/1234.jpg',0)
plt.figure(figsize=(15,10))
mean_filtered_img = meam_filter(imgrgb, kernel_size)
plt.subplot(2,2,1)
plt.imshow(mean_filtered_img)
plt.title("mean_filtered_img")
gaussian_filtered_img = gaussian_filter(imgrgb, kernel_size)
plt.subplot(2,2,2)
plt.imshow(gaussian_filtered_img)
plt.title("gaussian_filtered_img")
sharp_img = sharp(imgrgb)
plt.subplot(2,2,3)
plt.imshow(sharp_img, cmap='gray')
plt.title("sharp_img")
plt.show()
```
